```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Data distribution of numerical features
sns.pairplot(df, hue='HeartDisease')
plt.show()
```

```python
# Count of heart disease cases
plt.figure(figsize=(6,6))
ax = sns.countplot(x='HeartDisease', data=df)
plt.title('Count of Heart Disease Cases')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

plt.show()
```

```python
# Distribution of Age
sns.histplot(x='Age', data=df, kde=True, bins=20)
plt.title('Distribution of Age')
plt.show()
```

```python
# Gender distribution
sns.countplot(x='Sex', data=df)
counts = df['Sex'].value_counts()
for i, count in enumerate(counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.title('Gender Distribution')
plt.show()
```

```python
# Heart disease by gender
plt.figure(figsize=(8,6))
ax = sns.countplot(x='Sex', hue='HeartDisease', data=df)

# Add exact numbers on top of each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center', 
                 xytext = (0, 10), 
                 textcoords = 'offset points')

plt.title('Heart Disease by Gender')
plt.show()
```

```python
# Chest pain type distribution
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='ChestPainType', data=df)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')

plt.title('Chest Pain Type Distribution')
plt.show()
```

```python
# Effect of Age on Heart Disease
sns.boxplot(x='HeartDisease', y='Age', data=df)
plt.title('Effect of Age on Heart Disease')
plt.show()
```

```python
# Relationship between Resting Blood Pressure and Cholesterol
sns.scatterplot(x='RestingBP', y='Cholesterol', hue='HeartDisease', data=df)
plt.title('Relationship between Resting BP and Cholesterol')
plt.show()
```

```python
# Distribution of Maximum Heart Rate Achieved by Heart Disease Status
sns.histplot(x='MaxHR', hue='HeartDisease', data=df, kde=True, bins=20)
plt.title('Distribution of Maximum Heart Rate Achieved by Heart Disease')
plt.show()
```

```python
# Distribution of Oldpeak by Heart Disease Status
sns.histplot(x='Oldpeak', hue='HeartDisease', data=df, kde=True, bins=20)
plt.title('Distribution of Oldpeak by Heart Disease')
plt.show()
```

```python
# Comparison of Heart Disease Cases with and without ST-T wave abnormality
ax = sns.countplot(x='RestingECG_ST', hue='HeartDisease', data=df)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.title('Comparison of Heart Disease Cases with and without ST-T wave abnormality')
plt.show()
```

```python
# Effect of Fasting Blood Sugar on Heart Disease
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='FastingBS', hue='HeartDisease', data=df)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')

plt.title('Effect of Fasting Blood Sugar on Heart Disease')
plt.xlabel('Fasting Blood Sugar (mg/dl)')
plt.show()
```

```python
# Distribution of Serum Cholesterol Levels by Heart Disease Status
sns.histplot(x='Cholesterol', hue='HeartDisease', data=df, kde=True, bins=20)
plt.title('Distribution of Serum Cholesterol Levels by Heart Disease')
plt.xlabel('Serum Cholesterol (mm/dl)')
plt.show()
```

```python
# Distribution of Resting Blood Pressure by Heart Disease Status
sns.histplot(x='RestingBP', hue='HeartDisease', data=df, kde=True, bins=20)
plt.title('Distribution of Resting Blood Pressure by Heart Disease')
plt.xlabel('Resting Blood Pressure (mm Hg)')
plt.show()
```

```python
# Relationship between Age and Maximum Heart Rate Achieved by Heart Disease Status
sns.scatterplot(x='Age', y='MaxHR', hue='HeartDisease', data=df)
plt.title('Relationship between Age and MaxHR by Heart Disease')
plt.xlabel('Age')
plt.ylabel('MaxHR')
plt.show()
```
